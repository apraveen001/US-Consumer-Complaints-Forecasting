```python
# --- Handling Missing Values ---

# Drop columns with excessive missing values (>85%)
threshold = 0.85
missing_fraction = df.isnull().mean()
df = df.loc[:, missing_fraction < threshold]

# Fill missing categorical fields with 'Unknown'
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna('Unknown', inplace=True)

# Fill missing numerical columns (if any) with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# --- Save the cleaned dataset ---
output_path = "cleaned_consumer_complaints.csv"
df.to_csv(output_path, index=False)
print(f"Cleaned data saved to: {output_path}")
```
